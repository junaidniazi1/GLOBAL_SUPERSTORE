```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Optional settings
sns.set(style="whitegrid")
%matplotlib inline
```

## ð  Sales Overview

```python
# Total sales, profit, and average per order
print("Total Sales: ", df_merged['Sales'].sum())
print("Total Profit: ", df_merged['Profit'].sum())
print("Average Profit per Order: ", df_merged.groupby('Order ID')['Profit'].sum().mean())
```

## ð Time Series: Sales Over Time

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Order Date to datetime (in case it's not)
df_merged['Order Date'] = pd.to_datetime(df_merged['Order Date'])

# Create a proper datetime-based Month column as string (not Period)
df_merged['Order Month'] = df_merged['Order Date'].dt.to_period('M').astype(str)

# Group sales by Order Month
monthly_sales = df_merged.groupby('Order Month')['Sales'].sum().reset_index()

# Ensure 'Sales' is numeric (optional safety check)
monthly_sales['Sales'] = pd.to_numeric(monthly_sales['Sales'], errors='coerce')

# Plot the trend
plt.figure(figsize=(12,5))
sns.lineplot(x='Order Month', y='Sales', data=monthly_sales, marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```

## ð§© Segment and Region Analysis

```python
# Sales by Segment
segment_sales = df_merged.groupby('Segment')['Sales'].sum().reset_index()
sns.barplot(data=segment_sales, x='Segment', y='Sales')
plt.title("Sales by Customer Segment")
plt.show()

# Profit by Region
region_profit = df_merged.groupby('Region')['Profit'].sum().reset_index()
sns.barplot(data=region_profit, x='Region', y='Profit')
plt.xticks(rotation=45)
plt.title("Profit by Region")
plt.show()
```

## ðï¸  Product Performance

```python
# Top 10 profitable products
top_products = df_merged.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', figsize=(8,5), title="Top 10 Profitable Products")

# Top sub-categories by sales
plt.figure(figsize=(10,5))
sns.barplot(y='Sub-Category', x='Sales', data=df_merged, estimator=sum, order=df_merged.groupby('Sub-Category')['Sales'].sum().sort_values().index)
plt.title("Sales by Sub-Category")
```

```python
import pandas as pd

# If not already parsed
# df_merged['Order Date'] = pd.to_datetime(df_merged['Order Date'])

# ------------------------------
# ð¯ KPI 1: Total Sales
total_sales = df_merged['Sales'].sum()

# ð¯ KPI 2: Total Profit
total_profit = df_merged['Profit'].sum()

# ð¯ KPI 3: Profit Margin
profit_margin = (total_profit / total_sales) if total_sales else 0

# ð¯ KPI 4: Total Customers
total_customers = df_merged['Customer ID'].nunique()

# ð¯ KPI 5: Repeat Purchase Rate
customer_order_counts = df_merged.groupby('Customer ID')['Order ID'].nunique()
repeat_customers = (customer_order_counts > 1).sum()
repeat_purchase_rate = repeat_customers / total_customers if total_customers else 0

# ð¯ KPI 6: Churn Rate (monthly)
df_merged['Order Month'] = df_merged['Order Date'].dt.to_period('M')
monthly_customers = df_merged.groupby('Order Month')['Customer ID'].nunique()

# Calculate churn (optional: smooth with moving average)
churn_rate = monthly_customers.diff().fillna(0) / monthly_customers.shift(1).fillna(1)
avg_churn_rate = churn_rate.mean()

# ð¯ KPI 7: Average Order Value
total_orders = df_merged['Order ID'].nunique()
average_order_value = total_sales / total_orders if total_orders else 0

# ð¯ KPI 8: Average Quantity per Order
average_quantity_per_order = df_merged.groupby('Order ID')['Quantity'].sum().mean()

# ð¯ KPI 9: Return Rate (if 'Returned' column exists)
if 'Returned' in df_merged.columns:
    total_returns = df_merged[df_merged['Returned'] == 'Yes']['Order ID'].nunique()
    return_rate = total_returns / total_orders if total_orders else 0
else:
    return_rate = None

# ð¯ KPI 10: Customer Lifetime Value (Simple version)
# Avg revenue per customer * repeat purchase rate
avg_revenue_per_customer = total_sales / total_customers if total_customers else 0
clv = avg_revenue_per_customer * repeat_purchase_rate
```

```python
print("ð§¾ Total Sales: {:.2f}".format(total_sales))
print("ð° Total Profit: {:.2f}".format(total_profit))
print("ð Profit Margin: {:.2%}".format(profit_margin))
print("ð§ââï¸ Total Customers:", total_customers)
print("ð Repeat Purchase Rate: {:.2%}".format(repeat_purchase_rate))
print("ð Avg Monthly Churn Rate: {:.2%}".format(avg_churn_rate))
print("ðï¸ Average Order Value: {:.2f}".format(average_order_value))
print("ð¦ Avg Quantity per Order: {:.2f}".format(average_quantity_per_order))
if return_rate is not None:
    print("ð Return Rate: {:.2%}".format(return_rate))
print("ð§âð¼ Customer Lifetime Value (CLV): {:.2f}".format(clv))
```
